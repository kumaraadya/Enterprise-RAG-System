```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

manifest = pd.read_csv("../data/manifest.csv")
```

### Basic statistics

```python
print("Dataset Overview")
print(f"Total documents: {len(manifest)}")
print(f"Companies: {manifest['company'].nunique()}")
print(f"Date range: {manifest['filing_date'].min()} to {manifest['filing_date'].max()}")
```

### Documents per company

```python
print("\nDocuments per company:")
print(manifest['company'].value_counts())
```

### File sizes

```python
print("\nFile size statistics (KB):")
print(manifest['file_size_kb'].describe())
```

### Visualization

##### Chart 1: Documents per company
##### Chart 2: File size distribution 

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
manifest['company'].value_counts().plot(kind='barh', ax=axes[0])
axes[0].set_title('Documents per Company')
axes[0].set_xlabel('Count')

manifest['file_size_kb'].plot(kind='hist', bins=20, ax=axes[1])
axes[1].set_title('File Size Distribution')
axes[1].set_xlabel('Size (KB)')

plt.tight_layout()
plt.show()
```
